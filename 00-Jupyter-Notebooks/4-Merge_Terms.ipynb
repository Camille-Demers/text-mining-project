## **4. Merge Terms to prepare for Clustering**

```python
import pandas as pd
import glob
import os

base_path = '../05-transformation/'
liste = [base_path + f for f in os.listdir(base_path)]
  
df = pd.concat(map(pd.read_csv, liste), ignore_index=True).drop(columns=['Unnamed: 0', 'Fréquence (TF)', 'Fréquence documentaire (DF)', 'Terme formatté']).drop(columns=['isTaxoTerm'])
df = df[['Corpus', 'Terme', 'Structure syntaxique', 'isMeSHTerm', 'MeSHID', 'MesH_prefLabel_fr', 'MesH_prefLabel_en', 'LLR', \
    'TF (sklearn)', 'DF (sklearn)', 'TF-IDF', 'OkapiBM25']].rename(columns={"TF (sklearn)": "Term Frequency (TF)", "DF (sklearn)":"Document Frequency (DF)"})

df
```

```python
# Lister les corpus dans lesquels un même terme se retrouve
groups = df.groupby('Terme')['Corpus'].apply(list)

# groups store in a new 
# column called listvalues
df1 = groups.reset_index(name = 'Corpora')
df1

df = pd.merge(df1, df, on="Terme").drop(columns=['Corpus'])
df
```

```python
df['TF'] = df.groupby(['Terme'])['Term Frequency (TF)'].transform('sum').drop(columns=['Term Frequency (TF)'])
df['DF'] = df.groupby(['Terme'])['Document Frequency (DF)'].transform('sum').drop(columns=['Document Frequency (DF)'])
df['TF*IDF'] = df.groupby(['Terme'])['TF-IDF'].transform('max').drop(columns=['TF-IDF'])
df['OKapiBM25'] = df.groupby(['Terme'])['OkapiBM25'].transform('max').drop(columns=['OkapiBM25'])
df['Log Likelihood'] = df.groupby(['Terme'])['LLR'].transform('max').drop(columns=['LLR'])
df = df.drop_duplicates(['Terme'])

stop_path = '../04-filtrage/mwe_stopwords.txt'
with open(stop_path, encoding = 'utf-8') as f:
    stopwords = f.readlines()

for word in stopwords:
   df.drop(df[df['Terme'] == word.lower().strip()].index, inplace = True)
```

```python
from nltk.tokenize import RegexpTokenizer
import treetaggerwrapper
tagger = treetaggerwrapper.TreeTagger(TAGLANG='fr')

# lemmatisation
from french_lefff_lemmatizer.french_lefff_lemmatizer import FrenchLefffLemmatizer
lemmatizer = FrenchLefffLemmatizer()

file_path = '../04-filtrage/mapping_treeTagger_lefff.csv'

with open(file_path) as f:
   csv = pd.read_csv(f)

treeTag = [term for term in csv['TreeTagger'].tolist()] 
lefff = [term for term in csv['Lefff'].tolist()]

mapping = {term : lefff[treeTag.index(term)] for term in treeTag}


df['tokens'] = [[t.split('\t')[0] for t in tagger.tag_text(x)] for x in df['Terme']]
df['len_tokens'] = [len(x) for x in df['tokens'].tolist()]
df['tags'] = [x.split() for x in df['Structure syntaxique'].tolist()]
df['len_tags'] = [len(x) for x in df['tags'].tolist()]

df.drop(df[df['len_tokens'] != df['len_tags']].index, inplace = True)
df['tags_lefff'] = [[mapping[x] for x in t] for t in df['tags']]
```

```python
dic = df[['Terme', 'tokens', 'tags_lefff']].to_dict(orient='records')

for terme in dic:
    forme_lemmatisee = []
    for i in range(len(terme['tokens'])):
        tup = {'token': terme['tokens'][i], 'tag':terme['tags_lefff'][i], 'lemme' : terme['tokens'][i]}

        if lemmatizer.lemmatize(word=terme['tokens'][i], pos=terme['tags_lefff'][i]) == []:
            term_l = lemmatizer.lemmatize(word=terme['tokens'][i])

        elif type(lemmatizer.lemmatize(word=terme['tokens'][i], pos=terme['tags_lefff'][i])) == str:
            term_l  = lemmatizer.lemmatize(word=terme['tokens'][i], pos=terme['tags_lefff'][i])

        else:
            term_l = lemmatizer.lemmatize(word=terme['tokens'][i], pos=terme['tags_lefff'][i])[0][0]

        tup['lemme'] = term_l
        forme_lemmatisee.append(term_l)

    terme['Forme lemmatisée'] = " ".join(forme_lemmatisee).replace("' ", "'").replace("- ", "-").replace("( ", "(").replace(") ", ")")
```

```python
df = pd.merge(pd.DataFrame(dic).drop(columns=['tokens', 'tags_lefff']), df, on="Terme")
df = df[['Corpora', 'Terme', 'Forme lemmatisée', 'Structure syntaxique', 'isMeSHTerm', 'MeSHID', 'MesH_prefLabel_fr', 'MesH_prefLabel_en', 'Log Likelihood', 'TF', 'DF', 'TF*IDF',	'OKapiBM25']]
```

```python
# Extraction de termes existant dans la taxonomie 
df = df.assign(isTaxoTerm = False)
file_path = '../04-filtrage/default_taxo_labels.csv'

with open(file_path, 'r', encoding='utf-8') as f:
    default = pd.read_csv(f, sep=';')
    taxo_terms = list(dict.fromkeys([str(t).strip().lower() for t in default['Label'].tolist()]))

for t in taxo_terms:
    df.loc[((df['Terme'] == t) | \
            (df['MesH_prefLabel_fr'] == t) | \
            (df['MesH_prefLabel_en'] == t) | \
            (df['Forme lemmatisée'] == t)) , 'isTaxoTerm'] = True

df = df[['Corpora', 'Terme', 'Structure syntaxique','Forme lemmatisée','isMeSHTerm', 'MeSHID', 'MesH_prefLabel_fr', 'MesH_prefLabel_en', 'isTaxoTerm', 'Log Likelihood', \
   'TF', 'DF', 'TF*IDF', 'OKapiBM25']]

df[df['isTaxoTerm'] == True]
```

```python
print("N termes au départ = " + str(len(df)))
filtre = df[((df['TF'] > 30) & (df['DF'] > 20)) | (df['isMeSHTerm'] == True) | (df['isTaxoTerm'] == True)].sort_values(by = 'TF', ascending=False)
print("N termes après filtrage = " + str(len(filtre)))
```

```python
filtre.to_csv('../06-clustering/candidate_terms.csv')
```
